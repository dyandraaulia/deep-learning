```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_training = pd.read_csv('datasets/Google_Stock_Price_Train.csv')
training_set = dataset_training.iloc[:, 1:2].values
training_set
```

```python
dataset_training
```

```python
from sklearn.preprocessing import MinMaxScaler

sc = MinMaxScaler(feature_range=(0,1))
training_set_scaled = sc.fit_transform(training_set)
training_set_scaled
```

```python
X_train = []
y_train = []
for i in range(60, 1258):
    X_train.append(training_set_scaled[i-60:i, 0])
    y_train.append(training_set_scaled[i, 0])
```

```python
X_train
```

```python
y_train
```

```python
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
X_train
```

```python
y_train
```

```python
from keras import Sequential
from keras import layers
# from keras.layers import Dense, LSTM, Dropout
```

```python
regressor = Sequential()
#layer 1
regressor.add(layers.LSTM(units = 50, return_sequences= True, input_shape=(X_train.shape[1], 1)))
regressor.add(layers.Dropout(0.2))
```

```python
# layer 2
regressor.add(layers.LSTM(units=50, return_sequences=True))
regressor.add(layers.Dropout(0.2))
```

```python
# layer 3
regressor.add(layers.LSTM(units=50, return_sequences=True))
regressor.add(layers.Dropout(0.2))
```

```python
# layer 4
regressor.add(layers.LSTM(units=50))
regressor.add(layers.Dropout(0.2))
```

```python
regressor.add(layers.Dense(units=1))
```

```python
regressor.compile(optimizer='adam', loss='mean_squared_error')
```

```python
regressor.fit(X_train, y_train, epochs=100, batch_size=32)
```

```python
dataset_test = pd.read_csv('datasets/Google_Stock_Price_Test.csv')
real_stock_price = dataset_test.iloc[:,1:2].values
```

```python
dataset_total = pd.concat((dataset_training['Open'], dataset_test['Open']), axis= 0)
inputs = dataset_total[len(dataset_total) - len(dataset_test) - 60:].values
inputs = inputs.reshape(-1,1)
inputs = sc.transform(inputs)
```

```python
X_test = []

for x in range(60, 80):
    X_test.append(inputs[x-60:x, 0])
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
predicted_stock_price = regressor.predict(X_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```

```python
plt.plot(real_stock_price, color = 'red', label = 'Real Google Stock Price')
plt.plot(predicted_stock_price, color = 'blue', label = 'Predicted Google Stock Price')
plt.title('Google Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')
plt.legend()
plt.show()
```
